```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import os
```

Entrenamiento del Modelo de predicción

    Se usa el algoritmo Random Forest con los datos limpios.


```python
#Cargar los datos procesados
input_path="../data/clean/nba_clean.csv"
model_path="../models/model_test.pkl"

df = pd.read_csv(input_path)
df
```

Se utilizan los campos como variables de entrada (X) y el resultado del partido (WIN) como variable objetivo (y).

```python
# Seleccionar campos importantes
features = [
    "PTS_ROLL5", "REB_ROLL5", "AST_ROLL5",
    "FG_PCT_ROLL5", "FT_PCT_ROLL5", "PLUS_MINUS_ROLL5", "HOME"
]

X = df[features]
Y = df["WIN"]
```

```python
# Dividir en entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
# Crear y entrenar el modelo
model = RandomForestClassifier(
    n_estimators=150,
    max_depth=8,
    random_state=42,
    n_jobs=1
)
model.fit(X_train, Y_train)
```

```python
# Evaluacion
Y_pred = model.predict(X_test)
acc = accuracy_score(Y_test, Y_pred)
cm = confusion_matrix(Y_test, Y_pred)

print(f"Modelo entrenado con presición: {acc:.3f}")
print("Matriz de confusión:")
print(cm)
```
