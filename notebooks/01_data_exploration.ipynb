```python
from nba_api.stats.endpoints import leaguegamefinder
import pandas as pd
import os
```

RECOLECCIÓN DE LOS DATOS DE NBA_API

Columnas importantes:

    -TEAM_ABBREVIATION

    -PTS, REB, AST, FG_PCT, FT_PCT, PLUS_MINUS

    -GAME_DATE

    -WL (Win/Loss)

```python
# Crear carpeta si no existe
os.makedirs("../data/raw", exist_ok=True)

# Descargar los datos
games = leaguegamefinder.LeagueGameFinder().get_data_frames()[0]

# Guardar CSV
games.to_csv("../data/raw/nba_games.csv", index=False)
print("Archivo guardado en data/raw/nba_games.csv")
```

LIMPIEZA Y FEATURE

    Convierte la columna WL a 1 (Win) / 0 (Loss).

    Calcula medias móviles de los últimos 5 partidos por equipo (puntos, rebotes, asistencias, etc.).

    Agrega si el equipo juega en casa o fuera.

```python
#Cargar el csv a DataFRame
df = pd.read_csv("../data/raw/nba_games.csv")

#Convertir Win/Loss a 1/0
df["win"] = df["WL"].apply(lambda x: 1 if x == "W" else 0)

# Crear carpeta si no existe
os.makedirs("../data/clean", exist_ok=True)

df.to_csv("../data/clean/nba_clean.csv", index=False)
```
